```python
import math
import pdb # debugger

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
        
def get_percentage_index(l, p):
    return int(math.floor(len(l) / 100 * p))
```

```python
scaler = MinMaxScaler()
network = Sequential()
network.add(LSTM(units = 4, activation  = 'sigmoid', input_shape = (None, 1)))
network.add(Dense(units = 1))
network.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
data = pd.read_csv('https://stooq.com/q/d/l/?s=^aex&i=d')
data = data[get_percentage_index(data, 50):]
```

```python
data_transformed = data['Close'].values
data_transformed = [[x] for x in data_transformed]
data_transformed = scaler.fit_transform(data_transformed)
```

```python
training_index = get_percentage_index(data, 20)
data_testing = data_transformed[training_index:]
data_training = data_transformed[:training_index]
data_training_inputs = data_training[0:len(data_training) - 1]
data_training_targets = data_training[1:len(data_training)]
```

```python
network.fit(
    np.reshape(data_training_inputs, (len(data_training_inputs), 1, 1)),
    data_training_targets,
    verbose = 1,
    batch_size = 50,
    epochs = 500
)
```

```python
data_testing_prediction = network.predict(np.reshape(data_testing, (len(data_testing), 1, 1)))
```

```python
plot_training_inputs = [x[0] for x in scaler.inverse_transform(data_training_inputs)]
plot_training_targets = [x[0] for x in scaler.inverse_transform(data_training_targets)]

nones = [None for x in data_training]

plot_testing = nones + [x[0] for x in scaler.inverse_transform(data_testing)]
plot_prediction = nones + [x[0] for x in scaler.inverse_transform(data_testing_prediction)]

plot_testing = plot_testing[len(plot_testing) - 10:]
plot_prediction = plot_prediction[len(plot_prediction) - 10:]

#plt.plot(plot_training_inputs, color = 'purple', label = 'Training Prices')
#plt.plot(plot_training_targets, color = 'green', label = 'Training Targets')
plt.plot(plot_testing, color = 'magenta', label = 'Prices')
plt.plot(plot_prediction, color = 'cyan', label = 'Prediction')
plt.title('Prediction')
plt.xlabel('Days')
plt.ylabel('Price')
plt.legend()
plt.show()

data['Prediction'] = pd.Series(plot_prediction).values

data
```
